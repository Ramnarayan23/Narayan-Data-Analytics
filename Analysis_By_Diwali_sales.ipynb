```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'c:\Users\HCRMN\Downloads\Python_Diwali_Sales_Analysis-main\Python_Diwali_Sales_Analysis-main\Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df
```

```python
df.drop(columns=["Status", "unnamed1"] , inplace=True)   #dropping the columns
```

```python
df.drop(columns=["Satus", "unnamed1"], inplace=True, errors="ignore")   #inplace=True will make the changes in the original dataframe
```

```python
df.isna().sum()   #checking for missing values
```

```python
df.dropna(inplace=True)   #dropping the missing values
```

```python
import matplotlib.ticker as mticker
```

```python
Gender_amount=df.groupby("Gender")["Amount"].sum().reset_index()
```

```python
plt.figure(figsize=(8, 5))
ax=sns.barplot(x=Gender_amount["Amount"], y=Gender_amount["Gender"], palette="magma")
for index, value in enumerate(Gender_amount["Amount"]):
    plt.text(value + 200, index, f"₹{value}", fontsize=12, fontweight="bold")
plt.xlabel("Amount")
plt.ylabel("Gender")    
plt.title("Total Amount Spent by each Gender")  
plt.show()
```

```python
marital_st=sns.countplot(data=df, x="Marital_Status", hue="Gender")
for container in marital_st.containers:
    marital_st.bar_label(container)
plt.xlabel("Marital Status")
plt.ylabel("Gender")
plt.title("Marital Status of Customers")
plt.show()  
```

```python
totl_order = sns.countplot(data=df, x="Orders", hue="Gender")
for container in totl_order.containers:
    totl_order.bar_label(container)
plt.xlabel("Gender")
plt.ylabel("Number of Orders")
plt.title("Number of Orders Placed by Gender")  
plt.show()
```

```python
category_amount=df.groupby("Product_Category")["Amount"].sum().reset_index()
```

```python
category_amount=category_amount.sort_values(by="Amount", ascending=True)
```

```python
plt.figure(figsize=(10, 5))

# Create bar plot
ax = sns.barplot(x="Amount", y="Product_Category", data=category_amount, palette="viridis")

# Add labels to bars
for index, value in enumerate(category_amount["Amount"]):
    plt.text(value + 200, index, f"₹{value}", va="center", fontsize=12, fontweight="bold")

# Fix axis labels and title
plt.xlabel("Amount")
plt.ylabel("Product Category")
plt.title("Amount by Product Category")

plt.show()
```

```python

# Corrected Bar Plot: Total Amount by Marital Status
ax = sns.barplot(data=df, x="Marital_Status", y="Amount", estimator=sum, palette="viridis")

# Add Labels on Bars
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f")  # Format values with 2 decimals

# Labels & Title
plt.xlabel("Marital Status")
plt.ylabel("Total Amount")
plt.title("Total Amount Spent by Marital Status")

plt.show()
```
